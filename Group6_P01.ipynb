### Undersampling

```r
predictor_var <- bank_churns[,-13] 
response_var <- bank_churns$Exited
```

```r
levels(response_var) <- c('0', '1') 
```

```r
# Undersampling
# install.packages("unbalanced")
library("unbalanced")
undersampled_data <- ubBalance(predictor_var, 
                               response_var, 
                               type='ubUnder',         # Option for undersampling
                               verbose = TRUE)
```

```r
undersampled_combined <- cbind(undersampled_data$X, undersampled_data$Y)
undersampled_combined
names(undersampled_combined)[names(undersampled_combined) == "undersampled_data$Y"] <- "Exited"
levels(undersampled_combined$Exited) <- c('Exited', 'Present')
```

```r
ggplot(data = undersampled_combined, aes(fill = Exited))+
    geom_bar(aes(x = Exited))
```

### Oversampling

```r
oversampled_data <- ubBalance(predictor_var,  
                              response_var, 
                              type='ubOver',         
                              k = 0,                 # 0 => 50:50 split
                              verbose = TRUE)
```

```r
oversampled_combined <- cbind(oversampled_data$X,    
                               oversampled_data$Y)

names(oversampled_combined)[names(oversampled_combined) == "oversampled_data$Y"] <- "Exited"
levels(oversampled_combined$Exited) <- c('Present', 'Exited')
```

```r
ggplot(data = oversampled_combined, aes(fill = Exited))+
    geom_bar(aes(x = Exited))
```

## Log transformatoin for regular data

```r
skewness(bank_churns$CreditScore)
skewness(bank_churns$Age)
skewness(bank_churns$Tenure)
skewness(bank_churns$Balance)
skewness(bank_churns$NumOfProducts)
skewness(bank_churns$IsActiveMember)
skewness(bank_churns$EstimatedSalary)
```

If the skewness of the predictor variable is 0, the data is perfectly symmetrical

If the skewness of the predictor variable is less than -1 or greater than +1, the data is highly skewed

If the skewness of the predictor variable is between -1 and -0.5 or between +1 and +0.5 then the data is moderately skewed

If the skewness of the predictor variable is -0.5 and +0.5, the data is approximately symmetric

Therefore: Age is highly skewed, Number of Products are moderately skewed, and Balance, Creditscore and Tenure and estimated salary are approximately symmetric.
